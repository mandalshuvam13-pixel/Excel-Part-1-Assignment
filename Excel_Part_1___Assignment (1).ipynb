1. What is the difference between 'Paste' and 'Paste Special' in Excel? Briefly
explain with examples.
   - Let me briefly explain it to you:
   Paste: This is the default and most common pasting method. When you use 'Paste', Excel attempts to paste everything from the copied selection – including values, formulas, formatting, column widths, and comments – to the new location. It tries to replicate the source data as accurately as possible.
   
   Example: If you copy a cell containing a formula, a specific font color, and a border, 'Paste' will transfer the formula (which might adjust references), the font color, and the border to the destination cell.

   Paste Special (Ctrl+Alt+V or Home > Paste > Paste Special): This option provides a dialog box with various choices, allowing you to selectively paste specific attributes of the copied data. It gives you much finer control over the pasting process.

   Examples of 'Paste Special' options:
   - Values: Pastes only the calculated results of formulas, not the formulas themselves. This is useful when you want to remove dependencies on source cells or when you want to convert dynamic data into static data.
   
   Formats: Pastes only the formatting (e.g., font, color, borders, number formats) from the source cells, without the actual data or formulas.
   
   Formulas: Pastes only the formulas, without any formatting. Cell references in the formulas will adjust relative to the new location.
   
   Comments: Pastes only the comments attached to the cells.
   
   Column Widths: Pastes only the column widths from the source range to the destination range.
   
   All except borders: Pastes everything except the borders.
   
   Operations (Add, Subtract, Multiply, Divide): Performs a mathematical operation between the copied values and the values in the destination cells. For instance, you could add a copied value to a range of existing numbers.
   
   Transpose: Changes the orientation of the copied data from rows to columns, or vice-versa.

2. Describe the functions and usefulness of 'Freeze Panes' and 'Split Panes'
in Excel.
   - Freeze Panes
     - Function:
         'Freeze Panes' allows you to keep specific rows or columns visible as you scroll through the rest of your worksheet. This is particularly helpful when you have header rows or columns that contain important identifying information.
     
     - Usefulness:
       - When you have a large table of data, the first row often contains headers (e.g., 'Name', 'Date', 'Amount'). Freezing the top row keeps these headers always in view.
       
       - Similarly, if the first column contains unique identifiers (like customer IDs or product names), freezing it allows you to consistently see which record you're looking at, even if you scroll far to the right.
       
       - Used frequently in financial reports, inventory lists, customer databases, or any spreadsheet where consistent reference to row/column labels is essential.
       
       
    - Split Panes
      - Function:
         'Split Panes' divides your worksheet into separate, scrollable regions or panes. You can have two or four panes, and each pane can be scrolled independently.
      
      - Usefulness:
        - The primary use of splitting panes is to view and compare different, non-adjacent parts of the same worksheet simultaneously. For example, you might want to compare data from row 10 with data from row 500 without constantly scrolling back and forth.
        
        - In complex spreadsheets with multiple sections, splitting panes allows you to work on these sections concurrently.
        
        - Unlike 'Freeze Panes' which locks specific rows/columns, 'Split Panes' lets you scroll freely within each pane, making it more flexible for comparing any two areas.

4. What are logical functions in Excel? Provide examples of at least two
logical functions and their applications.
   - Logical functions in Excel are a category of functions that allow you to test for certain conditions and return a result based on whether those conditions are true or false. They are fundamental for decision-making within your spreadsheets, enabling you to create dynamic formulas that respond differently based on the data.

   Let's look at two common logical functions:
   1. IF Function:
     Purpose: The IF function checks a single condition and returns one value if the condition is TRUE, and another value if the condition is FALSE.
     
       Syntax: =IF(logical_test, value_if_true, value_if_false)
       
       logical_test: The condition you want to check (e.g., A1>10).
       
       value_if_true: What Excel should return if the logical_test is TRUE.
       
       value_if_false: What Excel should return if the logical_test is FALSE.
       
       Application Example: Grading System
       
       Imagine you have student scores in column B, and you want to assign 'Pass' or 'Fail' based on a passing score of 60.
       
       Formula: =IF(B2>=60, "Pass", "Fail")
       
       Explanation: If the score in B2 is greater than or equal to 60, the cell will display "Pass"; otherwise, it will display "Fail".
       
   2. AND Function
      
      Purpose: The AND function checks if all conditions provided are TRUE. It returns TRUE if all conditions are met, and FALSE if even one condition is not met.
      
      Syntax: =AND(logical1, [logical2], ...)
      
      logical1, logical2, ...: The conditions you want to check (up to 255 conditions).
      
      Application Example: Bonus Eligibility
      
      Suppose an employee gets a bonus if their 'Sales' (column C) are over $10,000 AND their 'Customer Satisfaction' (column D) is over 80%. (Note: AND is often nested within IF for practical outcomes).
      
      Formula: =IF(AND(C2>10000, D2>0.8), "Eligible for Bonus", "Not Eligible")
      
      Explanation: This formula first checks if both conditions (Sales > $10,000 AND Customer Satisfaction > 80%) are TRUE. If both are true, it returns "Eligible for Bonus"; otherwise, it returns "Not Eligible".

5. Discuss the purpose of 'XLOOKUP' and how it differs from the traditional
'VLOOKUP' function.
   - VLOOKUP (Vertical Lookup)
       Purpose: VLOOKUP is designed to look for a value in the first column of a table array and return a value from any cell on the same row from a column you specify.

   - XLOOKUP (Excel Lookup)
       Purpose: XLOOKUP is a modern and more flexible lookup function designed to overcome the limitations of VLOOKUP, HLOOKUP, and even INDEX/MATCH combinations. It searches for a value in one range (the lookup_array) and returns a corresponding item from a second range (the return_array).

6. Create a worksheet titled 'Employee Data' with columns: Name, Age,
Department. Add 5 rows of data.

   Format as follows:

   ● Bold and center-align the header row

   ● Apply a fill color

   ● Auto-fit column width

   (Include a screenshot showing your formatted table.)

`````python
import pandas as pd
from openpyxl import Workbook
from openpyxl.utils import get_column_letter
from openpyxl.styles import Font, Alignment, PatternFill

# Create a DataFrame with the employee data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [30, 24, 35, 29, 42],
    'Department': ['Sales', 'Marketing', 'Engineering', 'HR', 'Finance']
}
df = pd.DataFrame(data)

# Create a new Excel workbook and select the active sheet
workbook = Workbook()
ws = workbook.active
ws.title = "Employee Data"

# Write the DataFrame to the worksheet, starting from cell A1
# header=False to prevent pandas from writing headers again
# index=False to prevent pandas from writing DataFrame index
for r_idx, row in enumerate(df.values, start=2):
    for c_idx, value in enumerate(row, start=1):
        ws.cell(row=r_idx, column=c_idx, value=value)

# Manually write headers to row 1
headers = list(df.columns)
for c_idx, header_text in enumerate(headers, start=1):
    cell = ws.cell(row=1, column=c_idx, value=header_text)
    # Apply formatting to header row
    cell.font = Font(bold=True)
    cell.alignment = Alignment(horizontal='center')
    cell.fill = PatternFill(start_color="FFC7CEF0", end_color="FFC7CEF0", fill_type="solid") # Light blue fill

# Auto-fit column width and center-align data in columns
for col_idx, column in enumerate(ws.columns, start=1):
    max_length = 0
    column = [cell for cell in column] # Convert generator to list
    for cell in column:
        if cell.value is not None:
            # Calculate max length, considering header and data
            max_length = max(max_length, len(str(cell.value)))
            # Center-align all data cells (skip header as it's already done)
            if cell.row > 1:
                cell.alignment = Alignment(horizontal='center')
    # Set column dimension + a little extra padding
    adjusted_width = (max_length + 2)
    ws.column_dimensions[get_column_letter(col_idx)].width = adjusted_width

# Save the workbook
excel_file_name = 'Employee_Data.xlsx'
workbook.save(excel_file_name)

print(f"Excel file '{excel_file_name}' created successfully!")
`````

### **Initial Excel File (Before Changes)**

Let's first create a simple Excel file that we can use as our starting point. This file will be named `initial_data.xlsx`.

`````python
import pandas as pd
from openpyxl import Workbook

# Create a simple DataFrame
data = {
    'ID': [1, 2, 3, 4, 5, 6, 7, 8],
    'Product': ['Apple', 'Banana', 'Cherry', 'Date', 'Elderberry', 'Fig', 'Grape', 'Honeydew'],
    'Price': [1.0, 0.5, 2.5, 3.0, 1.2, 0.8, 1.5, 2.0]
}
df_initial = pd.DataFrame(data)

# Create a new workbook and write the DataFrame
workbook_initial = Workbook()
ws_initial = workbook_initial.active
ws_initial.title = "Product Data"

# Write headers
ws_initial.append(list(df_initial.columns))

# Write data rows
for row_data in df_initial.values:
    ws_initial.append(list(row_data))

# Save the initial workbook
initial_file_name = 'initial_data.xlsx'
workbook_initial.save(initial_file_name)

print(f"Initial Excel file '{initial_file_name}' created successfully!")
`````

### **Demonstrating Inserting Multiple Rows**

This code will open `initial_data.xlsx`, insert 3 new rows starting at row 3 (shifting existing rows down), and save it as `after_insert_rows.xlsx`.

`````python
from openpyxl import load_workbook

# Load the initial workbook
workbook_insert_rows = load_workbook('initial_data.xlsx')
ws_insert_rows = workbook_insert_rows.active

# Insert 3 rows starting at row 3 (below headers, shifting original row 3 (ID:3) to row 6)
# The insert_rows method inserts *before* the specified row index
# So, to insert 3 new rows before the data that was originally in row 3 (which is now row 3), we'll use 3
ws_insert_rows.insert_rows(idx=3, amount=3)

# Optionally, add some placeholder data to the new rows
ws_insert_rows.cell(row=3, column=2, value="NEW ROW 1")
ws_insert_rows.cell(row=4, column=2, value="NEW ROW 2")
ws_insert_rows.cell(row=5, column=2, value="NEW ROW 3")

# Save the modified workbook
after_insert_rows_file_name = 'after_insert_rows.xlsx'
workbook_insert_rows.save(after_insert_rows_file_name)

print(f"Excel file '{after_insert_rows_file_name}' (after inserting rows) created successfully!")
`````

`````python
import pandas as pd
from openpyxl import Workbook
from openpyxl.utils import get_column_letter
from openpyxl.styles import Alignment

# Create a DataFrame with sample employee data including department names
data = {
    'Employee ID': [101, 102, 103, 104, 105],
    'Name': ['John Doe', 'Jane Smith', 'Peter Jones', 'Alice Brown', 'Robert White'],
    'Department': ['Sales', 'Marketing', 'Sales', 'Human Resources', 'Marketing'],
    'Salary': [60000, 65000, 70000, 55000, 68000]
}
df_departments = pd.DataFrame(data)

# Create a new Excel workbook and select the active sheet
workbook_dept = Workbook()
ws_dept = workbook_dept.active
ws_dept.title = "Department Data"

# Write headers
ws_dept.append(list(df_departments.columns))

# Write data rows
for row_data in df_departments.values:
    ws_dept.append(list(row_data))

# Optional: Auto-fit column widths for better readability
for col_idx, column in enumerate(ws_dept.columns, start=1):
    max_length = 0
    for cell in column:
        if cell.value is not None:
            max_length = max(max_length, len(str(cell.value)))
        cell.alignment = Alignment(horizontal='center') # Center align all data
    adjusted_width = (max_length + 2)
    ws_dept.column_dimensions[get_column_letter(col_idx)].width = adjusted_width

# Save the initial workbook
departments_file_name = 'departments_initial.xlsx'
workbook_dept.save(departments_file_name)

print(f"Initial Excel file '{departments_file_name}' created successfully!")
`````

Create a small numerical dataset and apply the following functions:

`````python
from openpyxl import Workbook

# Create a new Excel workbook and select the active sheet
workbook_num = Workbook()
ws_num = workbook_num.active
ws_num.title = "Numerical Data"

# Create a small numerical dataset
data_values = [
    ["Value"],
    [10],
    [20],
    [15],
    [25],
    [30],
    [12],
    [18]
]

# Write the header and data values to the worksheet
for row_idx, row_data in enumerate(data_values, start=1):
    ws_num.cell(row=row_idx, column=1, value=row_data[0])

# Apply the functions (AVERAGE, MAX, MIN) using Excel formulas
# The dataset is in column A, from A2 to A8

# AVERAGE
ws_num.cell(row=10, column=1, value="Average:")
ws_num.cell(row=10, column=2, value="=AVERAGE(A2:A8)")

# MAX
ws_num.cell(row=11, column=1, value="Max:")
ws_num.cell(row=11, column=2, value="=MAX(A2:A8)")

# MIN
ws_num.cell(row=12, column=1, value="Min:")
ws_num.cell(row=12, column=2, value="=MIN(A2:A8)")

# Save the workbook
num_file_name = 'numerical_analysis.xlsx'
workbook_num.save(num_file_name)

print(f"Excel file '{num_file_name}' created successfully!")
`````

10. You're working with a dataset that contains missing values. As a Data
Scientist, explain how you'd detect and handle missing data using Excel.

Mention tools like:

● Go To Special

● ISBLANK

● COUNTBLANK

(Include a screenshot showing how blanks are identified or processed.)

`````python
import pandas as pd
from openpyxl import Workbook
from openpyxl.utils import get_column_letter
from openpyxl.styles import PatternFill

# Create a DataFrame with sample data including missing values
data_missing = {
    'ProductID': [101, 102, 103, 104, 105, 106, 107],
    'ProductName': ['Laptop', 'Mouse', None, 'Keyboard', 'Monitor', 'Webcam', None],
    'Category': ['Electronics', 'Electronics', 'Electronics', 'Peripherals', 'Electronics', 'Peripherals', 'Accessories'],
    'Price': [1200, 25, 75, None, 300, 50, 15],
    'Stock': [10, 50, 20, 15, None, 30, 5]
}
df_missing = pd.DataFrame(data_missing)

# Create a new Excel workbook and select the active sheet
workbook_missing = Workbook()
ws_missing = workbook_missing.active
ws_missing.title = "Missing Data Example"

# Write headers
ws_missing.append(list(df_missing.columns))

# Write data rows, replacing pandas 'None' with actual empty cells in Excel
for row_data in df_missing.values:
    ws_missing.append([None if pd.isna(cell) else cell for cell in row_data])

# Auto-fit column widths for better readability
for col_idx, column in enumerate(ws_missing.columns, start=1):
    max_length = 0
    for cell in column:
        if cell.value is not None:
            max_length = max(max_length, len(str(cell.value)))
    adjusted_width = (max_length + 2)
    ws_missing.column_dimensions[get_column_letter(col_idx)].width = adjusted_width

# Save the workbook
missing_data_file_name = 'missing_data_example.xlsx'
workbook_missing.save(missing_data_file_name)

print(f"Excel file '{missing_data_file_name}' created successfully!")
`````
